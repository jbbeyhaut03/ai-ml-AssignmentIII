This notebook demonstrates a basic example of how to load and process part of the Goodreads spoilers dataset using JSON. Specifically, it walks through loading user IDs and book IDs from the dataset.

**Important Notes:**

```python
import gzip
import json
```

```python
n_review = 0
max_number_review = 5000   # change this number to load more reviews

with gzip.open('goodreads_reviews_spoiler.json.gz') as fin:
    book_set = []
    user_set = []   # in this example we just extract a list of book ids and user ids

    for l in fin:
        d = json.loads(l)      # read the review data
        
        # add book id and user id to a set
        book_set.append(d['book_id'])
        user_set.append(d['user_id'])
        
        if n_review == max_number_review:
            break
```
